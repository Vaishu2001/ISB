```
import seaborn as sns
import pandas as pd
```

```
odi=pd.read_csv("/content/CricinfoICC ODI Championship.csv")
odi
```

Create a new dataframe for womens's odi

```
wodi=pd.read_csv("/content/CricinfoICC Women's ODI Team Rankings.csv")
wodi
```

```
def trim(x):
  a=x[:3]
  return a
```

```
wodi['team']=odi['team'].apply(trim)
wodi
```

```
odi['team']=odi['team'].apply(trim)
odi
```

Let us add a new column to wodi

```
wodi['Gender']='Female'
wodi
```

```
odi['Gender']="Male"
odi
```

To add 2 dataframes

```
comb=pd.concat([odi,wodi],ignore_index=True)
comb
```

```
f,ax=plt.subplots(figsize=(20,15))

ax=sns.barplot('team','matches',hue='Gender',data=comb)
```

```
f,ax=plt.subplots(figsize=(20,15))

ax=sns.scatterplot('team','matches',hue='Gender',data=comb)
```
