```
import matplotlib.pyplot as plt
import pandas as pd
```

converting csv file to dataframe

```

df=pd.read_csv("/content/kasaragodaugust.csv")
df1=pd.read_csv("/content/mangaloreaugust.csv")
```

```
df1
```

To draw a line chart:

```

plt.plot(df['Date'],df['High temparature'])
plt.title("Date vs High temparature",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("High temparature in degree celsius",fontsize=15)
```

```
plt.plot(df['Date'],df['Low temparature'])
plt.title("Date vs Low temparature",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("Low temparature in degree celsius",fontsize=15)
```

```
plt.plot(df['Date'],df['Precipitation'])
plt.title("Date vs Precipitation",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("Precipitation in mm",fontsize=15)
```

```
plt.subplot(2,1,1)

plt.plot(df['Date'],df['High temparature'])
plt.title("Date vs High temparature",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("High temparature in degree celsius",fontsize=15)
plt.show()
plt.subplot(2,1,2)
plt.plot(df['Date'],df['Low temparature'])
plt.title("Date vs Low temparature",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("Low temparature in degree celsius",fontsize=15)
plt.show()
```

```
plt.subplot(2,1,1)

plt.plot(df['Date'],df['High temparature'])
plt.title("Kasaragod",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("High temparature in degree celsius",fontsize=15)
plt.show()
plt.subplot(2,1,2)
plt.plot(df1['Date'],df1['High temparature'])
plt.title("Mangalore",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("High temparature in degree celsius",fontsize=15)
plt.show()
```
